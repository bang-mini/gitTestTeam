### requests 라이브러리 사용법

```python
# 라이브러리 불러오기 
import requests as req
# 서버에게 데이터를 요청 및 응답받기 위한 라이브러리
```

```python
res = req.get('https://www.naver.com/')
# <Response [200]>번대 : 성공적으로 요청 및 응답이 이루어졌다는 의미
# <Response [400]>번대 : 클라이언트(요청) 부분에서 오류가 발생했다는 의미
# <Response [500]>번대 : 서버(응답) 부분에서 오류가 발생했다는 의미
```

```python
res.text
```

### BeautifulSoup 라이브러리

```python
from bs4 import BeautifulSoup as bs
# 컴퓨터가 이해할 수 있는 객체 형태로 변환해주는 라이브러리
```

```python
# BeautifulSoup 객체화
# bs(parsing할 html 문서, 'parsing 방법')
# lxml : 빠르고 관대하다는 장점
soup = bs(res.text, 'lxml')
soup
```

```python
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query='
```

- 네이버페이지에서 정보를 가져와보자
- 뉴스, 카페, 블로그등과 같은 글자를 수집해보자


```python
res = req.get(url)
res
```

```python
res.text
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
# 태그의 위치 찾는 방법
# 1. 내가 가져오고 싶은 contents 위에 마우스 우클릭 -> 검사
# 2. F12(개발자모드) 좌측 상단에 마우스 아이콘 클릭 -> 내가 가져오고 싶은 contents 클릭
```

```python
# a태그 가져오기
soup.select('a')
# 페이지에 있는 모든 a태그가 전부 가져와졌다!
# select() : 페이지에 있는 태그를 모두(복수형태 : list) 가져왔다.
```

```python
soup.select_one('a')
# select_one() : 페이지에 있는 태그 중에서 가장 첫 번째 요소를 하나만(단수형태) 가져왔다
```

```python
# a태그 중에서 tab이라는 class 선택자들만 가져오는 방법
soup.select('a.tab')[0]
# 요소에서 태그를 없애고 contents 만 가져오겠다
soup.select('a.tab')[0].text
```
